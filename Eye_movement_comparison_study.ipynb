```python
from scipy.stats import mannwhitneyu

# Data for group A
group_a_data = {
    "Average Duration of Whole Fixations": [134, 297.3, 388, 362.7, 891.7],
    "Number of Whole Fixations": [58.7, 87, 70.8, 65.5, 33.2],
    "Number of Saccades": [17, 66.3, 63.3, 62.2, 28.2],
    "Average Peak Velocity of Saccades": [119.08, 188.433, 217.031, 151.444, 123.449]
}

# Data for group B
group_b_data = {
    "Average Duration of Whole Fixations": [253.8, 274.2, 261.8, 300.5, 386.4],
    "Number of Whole Fixations": [86.7, 91.2, 65.4, 68.7, 55.7],
    "Number of Saccades": [70.6, 82.4, 54.1, 55.4, 45.8],
    "Average Peak Velocity of Saccades": [158.914, 202.409, 109.24, 184.769, 202.484]
}

# Perform Mann-Whitney U test for each feature
for feature in group_a_data.keys():
    statistic, p_value = mannwhitneyu(group_a_data[feature], group_b_data[feature])
    print(f"Mann-Whitney U test for {feature}:")
    print(f"Statistic: {statistic}")
    print(f"P-value: {p_value}")
    if p_value < 0.05:
        print("The difference is statistically significant (p < 0.05)")
    else:
        print("The difference is not statistically significant (p >= 0.05)")
    print()
```

### Box Plots and Violin plots

**Parts-Box Plots**

Median (Q2): The line inside the box represents the median of the dataset, which is the middle value when the data is sorted.

Quartiles (Q1 and Q3): The box represents the interquartile range (IQR), which is the middle 50% of the data. The bottom edge of the box (Q1) represents the 25th percentile, and the top edge (Q3) represents the 75th percentile.

Whiskers: The whiskers extend from the edges of the box to indicate the range of the data. They typically represent 1.5 times the IQR beyond the quartiles. Any data points outside the whiskers are considered outliers.

Outliers: Individual data points that fall outside the whiskers are plotted as individual points. They are often represented as dots or circles.

Interpreting Spread: The length of the box and the whiskers provides an indication of the spread or variability of the data. Longer whiskers indicate a larger spread of the data, while shorter whiskers suggest a more clustered distribution.

Skewness: The orientation of the box (vertical or horizontal) can indicate the skewness of the data distribution. A symmetrical distribution will have a box plot with a vertical orientation, while a skewed distribution may have a horizontal orientation.






Box plots show the spread of the data and where the middle 50% of the data falls. In this case, the spread of the data is the duration of the fixations. The horizontal line in the middle of the box is the median duration, and the box shows the range of the middle 50% of the data. The lines extending from the box (whiskers) show the rest of the data up to 1.5 times the interquartile range (IQR) from the top and bottom of the box. If there are data points that fall outside of the whiskers, they are plotted as individual circles.

**Parts-Violin Plots**

Violin body: This is the central part of the plot and represents the distribution of the data. It is typically shaped like a violin or a kernel density plot, showing the probability density of the data at different values.

Width: The width of the violin body indicates the density of the data at different values. A wider section indicates higher density of data points, while a narrower section indicates lower density.

Interquartile range (IQR): Similar to a box plot, the violin plot often includes lines or bars within the violin body to represent the interquartile range of the data. This provides information about the spread of the data.

Median line: A line or marker within the violin body represents the median of the data, providing a measure of central tendency.

Ends or Extremes: In some violin plots, the ends of the violin may include lines or markers to indicate the minimum and maximum values of the data, or they may extend to show the entire range of the data.
